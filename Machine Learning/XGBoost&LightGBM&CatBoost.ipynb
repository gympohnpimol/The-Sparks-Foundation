### make pipeline 

```python
#encoder & transform columns
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.impute import SimpleImputer

#pipeline & cross validation
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
df.dtypes
```

```python
numeric_features = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 
                    'Outlet_Establishment_Year']

numeric_transformer = Pipeline(steps= [('poly', PolynomialFeatures(degree=2)),
                                      ('scaler', StandardScaler())])
```

```python
categorical_features = ['Item_Identifier', 'Item_Fat_Content', 'Item_Type',
                       'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', LinearRegression())])
```

```python
X = df.drop('Item_Outlet_Sales', axis=True)
X
```

```python
y = df['Item_Outlet_Sales']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_train, y_train)
```

```python
y_pred =clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
