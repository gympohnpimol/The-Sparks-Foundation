## Model Design

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

plt.figure(figsize  = (15,15))
tree.plot_tree(clf,filled = True,rounded = True,proportion = True,node_ids = True)
plt.show()
```
